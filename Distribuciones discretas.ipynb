```python
from scipy.stats import binom, poisson, geom, hypergeom, nbinom, bernoulli, randint

# Ejemplo 1: Distribución Binomial
n, p = 10, 0.5
x_binom = 5
prob_binom = binom.cdf(x_binom - 1, n, p)

# Ejemplo 2: Distribución Poisson
lambda_poisson = 3
x_poisson = 4
prob_poisson = poisson.cdf(x_poisson - 1, lambda_poisson)

# Ejemplo 3: Distribución Geométrica
p_geom = 0.2
x_geom = 3
prob_geom = geom.cdf(x_geom - 1, p_geom)

# Ejemplo 4: Distribución Hipergeométrica
N, K, n = 20, 7, 12  # Población total, número de éxitos en la población, tamaño de muestra
x_hyper = 3
prob_hyper = hypergeom.cdf(x_hyper - 1, N, K, n)

# Ejemplo 5: Distribución Binomial Negativa
r, p = 5, 0.5  # Número de éxitos, probabilidad de éxito
x_nbinom = 10
prob_nbinom = nbinom.cdf(x_nbinom - 1, r, p)

# Ejemplo 6: Distribución de Bernoulli
p = 0.3
x_bernoulli = 1
prob_bernoulli = bernoulli.cdf(x_bernoulli - 1, p)

# Ejemplo 7: Distribución Uniforme Discreta
low, high = 1, 10
x_randint = 5
prob_randint = randint.cdf(x_randint - 1, low, high)

# Imprimir resultados
print(f"Distribución Binomial P(X < {x_binom}): {prob_binom}")
print(f"Distribución Poisson P(X < {x_poisson}): {prob_poisson}")
print(f"Distribución Geométrica P(X < {x_geom}): {prob_geom}")
print(f"Distribución Hipergeométrica P(X < {x_hyper}): {prob_hyper}")
print(f"Distribución Binomial Negativa P(X < {x_nbinom}): {prob_nbinom}")
print(f"Distribución de Bernoulli P(X < {x_bernoulli}): {prob_bernoulli}")
print(f"Distribución Uniforme Discreta P(X < {x_randint}): {prob_randint}")
```
