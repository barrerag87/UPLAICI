En estadística, las funciones de probabilidad describen cómo se distribuyen las probabilidades de los diferentes resultados de una variable aleatoria. Estas funciones se clasifican en dos categorías principales: **función de probabilidad discreta** y **función de probabilidad continua**. 

### Función de Probabilidad Discreta

La **función de probabilidad discreta (PMF, por sus siglas en inglés)** se aplica a variables aleatorias discretas. Esto es, para variables que toman valores contables, como los enteros. La PMF asigna una probabilidad a cada posible valor de la variable aleatoria.

**Ejemplo: Distribución Binomial**

La distribución binomial es un ejemplo clásico de una variable aleatoria discreta, donde \( p \) es la probabilidad de éxito en un ensayo y $n$ es el número total de ensayos.



La distribución normal o gaussiana es quizás la más conocida de las distribuciones continuas, caracterizada por dos parámetros: la media $\mu$ y la desviación estándar $\sigma$.




```python

from scipy.stats import binom
import matplotlib.pyplot as plt

# Parámetros de la distribución
n = 10  # número de ensayos
p = 0.5  # probabilidad de éxito en cada ensayo

# Crear una variable aleatoria binomial
rv = binom(n, p)

# Valores posibles de X
x = list(range(n + 1))

# PMF para cada valor de X
pmf_vals = rv.pmf(x)

# Graficar la PMF
plt.bar(x, pmf_vals, color='blue')
plt.title('Función de probabilidad discreta de una Binomial(n=10, p=0.5)')
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad')
plt.show()
```



### Función de Probabilidad Continua

La **función de probabilidad continua (PDF, por sus siglas en inglés)** se utiliza en el contexto de variables aleatorias continuas, que toman un número infinito de valores posibles. En el caso de las variables continuas, hablamos de la densidad de probabilidad, ya que la probabilidad de obtener un valor exacto siempre es cero; en su lugar, se considera la probabilidad de caer dentro de un intervalo específico.

**Ejemplo: Distribución Normal**

```python

from scipy.stats import norm
import numpy as np

# Parámetros de la distribución
mu = 0     # media
sigma = 1  # desviación estándar

# Crear una variable aleatoria normal
rv = norm(mu, sigma)

# Rango de valores para x
x = np.linspace(-3*sigma, 3*sigma, 100)

# PDF para cada valor de x
pdf_vals = rv.pdf(x)

# Graficar la PDF
plt.plot(x, pdf_vals, color='red')
plt.title('Función de probabilidad continua de una Normal(μ=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.show()
```

### Distribución de Poisson

La **distribución de Poisson** es útil para modelar el número de eventos en un intervalo fijo de tiempo o espacio, dado un promedio constante de eventos por intervalo, conocido como $\lambda$.

**Implementación en Python:**

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parámetro de la distribución
lambda_ = 3  # tasa promedio de eventos por intervalo

# Valores posibles de X
x = list(range(0, 10))  # Rango razonable para visualizar

# PMF para cada valor de X
pmf_vals = poisson.pmf(x, lambda_)

# Graficar la PMF
plt.bar(x, pmf_vals, color='orange')
plt.title('Función de probabilidad discreta de una Poisson(λ=3)')
plt.xlabel('Número de eventos')
plt.ylabel('Probabilidad')
plt.show()
```

### Distribución Exponencial

La **distribución exponencial** describe el tiempo entre eventos en un proceso Poisson. Es una distribución continua con parámetro $ \lambda $ que es la tasa promedio de eventos por intervalo.

**Implementación en Python:**

```python

from scipy.stats import expon
import numpy as np

# Parámetro de la distribución
lambda_ = 1  # tasa promedio de eventos

# Rango de valores para x
x = np.linspace(0, 5, 100)

# PDF para cada valor de x
pdf_vals = expon.pdf(x, scale=1/lambda_)

# Graficar la PDF
plt.plot(x, pdf_vals, color='purple')
plt.title('Función de probabilidad continua de una Exponencial(λ=1)')
plt.xlabel('Tiempo')
plt.ylabel('Densidad de probabilidad')
plt.show()


```

**Ejercicio:** Calcula la probabilidad de que el tiempo hasta el próximo evento sea de más de 1 unidad de tiempo, dado que la tasa promedio de eventos es de 1 por unidad de tiempo.
